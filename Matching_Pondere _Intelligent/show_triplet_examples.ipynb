```python
# show_triplet_examples.py
"""
Affiche quelques exemples de triplets unifiÃ©s pour validation manuelle.
IdÃ©al pour vÃ©rifier la qualitÃ© du matching.
"""

import json
from pathlib import Path

def main():
    output_dir = Path("output")
    with open(output_dir / "unified_triplets.json", "r", encoding="utf-8") as f:
        triplets = json.load(f)

    if not triplets:
        print("âŒ Aucun triplet trouvÃ©.")
        return

    print(f"ðŸ” Affichage de 5 exemples parmi {len(triplets)} triplets unifiÃ©s :\n")

    # Afficher les 5 premiers (ou moins s'il y en a moins)
    for i, triplet in enumerate(triplets[:5]):
        print(f"{'='*80}")
        print(f"ðŸ”¸ Exemple {i+1} | ID: {triplet['unified_id']} | Score: {triplet['score']:.4f}")
        print(f"{'='*80}")

        for profile in triplet["profiles"]:
            platform = profile["platform"]
            name = profile.get("fullName", "â€”")
            username = profile.get("username", "â€”")
            email = profile.get("email", "â€”")
            bio = (profile.get("bio") or profile.get("headline") or "â€”")[:120]  # tronquer

            # Nettoyer la bio pour l'affichage
            bio = bio.replace('\n', ' ').replace('\r', ' ')

            print(f"\nâž¡ï¸ {platform.capitalize()}:")
            print(f"   Nom complet : {name}")
            print(f"   Username    : {username}")
            print(f"   Email       : {email}")
            print(f"   Bio/Headline: {bio}...")
        print("\n")

    # Optionnel : afficher un exemple alÃ©atoire avec score Ã©levÃ©
    if len(triplets) > 5:
        best = max(triplets, key=lambda t: t["score"])
        if best["score"] > triplets[0]["score"]:  # s'il n'est pas dÃ©jÃ  dans les 5
            print(f"ðŸŒŸ Meilleur triplet (score = {best['score']:.4f}):")
            profile = best["profiles"][0]
            print(f"   Ex: {profile['platform'].capitalize()} â€” {profile.get('fullName', 'â€”')} (@{profile.get('username', 'â€”')})")

if __name__ == "__main__":
    main()
```
