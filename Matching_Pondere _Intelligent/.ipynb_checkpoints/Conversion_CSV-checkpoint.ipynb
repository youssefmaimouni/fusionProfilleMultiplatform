```python
import json
import csv
import os

# === 1. Dossier contenant les fichiers JSON ===
dossier_data = "data"

# === 2. Fonction de conversion JSON → CSV ===
def convertir_json_en_csv(fichier_json):
    nom_csv = os.path.splitext(fichier_json)[0] + ".csv"
    
    # Charger les données JSON
    with open(fichier_json, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    # Si c’est un dictionnaire unique, le transformer en liste
    if isinstance(data, dict):
        data = [data]
    
    # Extraire toutes les clés pour les en-têtes du CSV
    cles = set()
    for item in data:
        cles.update(item.keys())
    cles = list(cles)

    # Écrire le fichier CSV
    with open(nom_csv, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=cles)
        writer.writeheader()
        writer.writerows(data)
    
    # Afficher le résultat
    taille = os.path.getsize(nom_csv) / 1024
    print(f"✅ {os.path.basename(fichier_json)} → {os.path.basename(nom_csv)} ({taille:.2f} Ko)")

# === 3. Parcourir le dossier 'data' et convertir les fichiers JSON ===
for fichier in os.listdir(dossier_data):
    if fichier.endswith(".json"):
        chemin_fichier = os.path.join(dossier_data, fichier)
        convertir_json_en_csv(chemin_fichier)

print("\n✅ Conversion terminée pour tous les fichiers JSON du dossier 'data/' !")
```
