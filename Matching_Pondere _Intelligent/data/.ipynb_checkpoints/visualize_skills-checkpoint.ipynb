```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from collections import Counter

# ----------------------------
# CONFIGURATION GLOBALE
# ----------------------------
plt.rcParams['font.size'] = 10
sns.set_style("whitegrid")

# ----------------------------
# CHARGEMENT DES DONNES
# ----------------------------
df = pd.read_csv("global_tech_talents_morocco.csv")
df['is_based_in_morocco'] = df['is_based_in_morocco'].astype(bool)

# ----------------------------
# FONCTION DE NORMALISATION DES VILLES
# ----------------------------
def normalize_location(loc):
    if pd.isna(loc) or str(loc).strip() == "":
        return ""
    loc = str(loc).strip()
    loc_lower = loc.lower()
    if "morocco" in loc_lower or "maroc" in loc_lower:
        city_part = re.split(r'[,;]', loc, maxsplit=1)[0].strip()
        city_part = re.sub(r'prefecture of\s*', '', city_part, flags=re.IGNORECASE)
        city_part = re.sub(r'-.*$', '', city_part).strip()
        if city_part.lower() in ["morocco", "maroc", "morocco ", ""]:
            return "Morocco"
        return city_part.title()
    return loc

df['location_clean'] = df['location'].apply(normalize_location)

morocco_df = df[df['is_based_in_morocco']].copy()
morocco_df['city'] = morocco_df['location_clean']

# ----------------------------
# PRPARATION DES DONNES POUR LES GRAPHIQUES
# ----------------------------

# 1. Plateformes
platform_data = df['primary_platform'].value_counts()

# 2. Rles (nettoyage)
def clean_role(role):
    if pd.isna(role) or str(role).strip() == "":
        return ""
    r = str(role).lower()
    r = re.sub(r'\s*at\s.*', '', r)
    r = re.sub(r'[^a-z\s]', ' ', r)
    return r.strip().title()

df['clean_role'] = df['current_role'].apply(clean_role)
roles = df[df['clean_role'] != '']['clean_role'].value_counts().head(10)

# 3. Comptences
all_skills = []
for sk in df['skills'].dropna():
    for s in str(sk).split(";"):
        s_clean = s.strip()
        if s_clean and s_clean not in ["", "N/A"]:
            all_skills.append(s_clean)

top_skills_series = pd.Series(Counter(all_skills).most_common(15))
if not top_skills_series.empty:
    top_skills = pd.Series(dict(top_skills_series.tolist())).sort_values()
else:
    top_skills = pd.Series([], dtype='float64')

# 4. Villes au Maroc
city_counts = morocco_df['city'].value_counts().head(10).sort_values()

# 5. Longueur moyenne des textes
text_lengths = {
    'Bio': df['bio'].fillna("").str.len().mean(),
    'Projets': df['projects_summary'].fillna("").str.len().mean(),
    'Comptences': df['skills'].fillna("").str.len().mean()
}
text_df = pd.Series(text_lengths)

# ----------------------------
# CRATION DES VISUALISATIONS (5 graphiques)
# ----------------------------
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Masquer le premier subplot (axes[0,0]) car on ne veut plus la gographie
axes[0, 0].axis('off')

# Titre principal
fig.suptitle(
    "Analyse des Talents Tech au Maroc  Donnes issues de GitHub, LinkedIn et Twitter",
    fontsize=14,
    weight='bold'
)

# 1. Plateformes  axes[0,1]
platform_data.plot(kind='bar', ax=axes[0, 1], color='#50C878')
axes[0, 1].set_title("Rpartition par plateforme principale")
axes[0, 1].set_ylabel("Nombre de profils")
axes[0, 1].tick_params(axis='x', rotation=45)

# 2. Top rles  axes[0,2]
if not roles.empty:
    roles.sort_values().plot(kind='barh', ax=axes[0, 2], color='#2196F3')
axes[0, 2].set_title("Top 10 des rles professionnels")
axes[0, 2].set_xlabel("Nombre de profils")

# 3. Top comptences  axes[1,0]
if not top_skills.empty:
    top_skills.plot(kind='barh', ax=axes[1, 0], color='#FF6F00')
axes[1, 0].set_title("Top 15 des comptences techniques")
axes[1, 0].set_xlabel("Nombre de mentions")

# 4. Villes au Maroc  axes[1,1]
if not city_counts.empty:
    city_counts.plot(kind='barh', ax=axes[1, 1], color='#9C27B0')
axes[1, 1].set_title("Top 10 des villes au Maroc")
axes[1, 1].set_xlabel("Nombre de profils")

# 5. Longueur moyenne des descriptions  axes[1,2]
text_df.plot(kind='bar', ax=axes[1, 2], color='#FF9800')
axes[1, 2].set_title("Longueur moyenne des champs textuels (caractres)")
axes[1, 2].set_ylabel("Caractres")
axes[1, 2].tick_params(axis='x', rotation=45)

# Ajustement final
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("tech_talents_analysis.png", dpi=300, bbox_inches='tight')
plt.show()
```
